```python
from textblob import TextBlob
```

```python
# polarity classification of the review is postive or negative 
# calculate polarity of reviews
def get_polarity(text):
    textblob=TextBlob(str(text.encode("utf-8"))) # هياخد العمود التيكست بتاع الريفيوز ويعملها انكود معين عشان يقدر يحوله لبولارتي
    pol=textblob.sentiment.polarity # make polarity of encoded reviews from -1 to 1
    return pol

#apply the function
df["polarity"]=df["verified_reviews"].astype(str).apply(get_polarity)
```

```python
# subjectivity الموضوعية هي تقييم هل الريفيو بيتكلم عن المنتج فعلا وبلا انحياز ولا لا وبتبقا رينج من 0ل1
def get_subjectivity(subject):
    textblob=TextBlob(str(subject.encode('utf-8')))
    sub=textblob.sentiment.subjectivity
    return sub
df["subjectivity"]=df["verified_reviews"].astype(str).apply(get_subjectivity)
```

```python
# code for counting nouns in reviews
pos_dic = {
    'noun': ['NN', 'NNS', 'NNP', 'NNPS'],
    'pron': ['PRP', 'PRP$', 'WP', 'WP$'],
    'verb': ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ'],
    'adj': ['JJ', 'JJR', 'JJS'],
    'adv': ['RB', 'RBR', 'RBS', 'WRB']
}

def pos_check(x, flag):
    cnt = 0
    try:
        wiki = TextBlob(x)
        for tup in wiki.tags:
            ppo = tup[1]
            if ppo in pos_dic[flag]:
                cnt += 1
    except Exception as e:
        print(f"Error processing text: {x}")
        print(f"Exception: {e}")
    return cnt

df['noun_count'] = df['verified_reviews'].astype(str).apply(lambda x: pos_check(x, 'noun'))
print(df)
```

```python
def punctuation_removal(messy_str):
    if isinstance(messy_str, str):
        clean_list = [char for char in messy_str if char not in string.punctuation]
        clean_str = ''.join(clean_list)
        return clean_str
    else:
        return str(messy_str)

df['verified_reviews'] = df['verified_reviews'].apply(punctuation_removal)
```

```python
# lets make a function to remove Numbers from the reviews
import re
def drop_numbers(list_text):
    list_text_new = []
    for i in list_text:
        if not re.search('\d', i): # searh of numbers in string \d
            list_text_new.append(i)
    return ''.join(list_text_new)

df['verified_reviews'] = df['verified_reviews'].apply(drop_numbers)
```

```python
# lets create a function to remove accented characters
import unicodedata # Import the unicodedata module        é

def remove_accented_chars(text):
    new_text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    return new_text

# lets apply the function
df['verified_reviews'] = df.apply(lambda x: remove_accented_chars(x['verified_reviews']), axis = 1)
```

```python
#Removal of special characters
def remove_special_characters(text):
    pat=r'[^a-zA-z0-9]'
    return re.sub(pat, ' ',text)

df["verified_reviews"]=df.apply(lambda x: remove_special_characters(x['verified_reviews']), axis=1)
```

```python
# cleaning the texts
# importing the libraries for Natural Language Processing

import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
## Cleaning the Data

corpus = []

for i in range(0, 3150):
    review = re.sub('[^a-zA-Z]', ' ', df['verified_reviews'][i])  ## Removing all Unecessary items
    review = review.lower()                                         ## Converting into Lower Case
    review = review.split()
    ps = PorterStemmer()                                            ## Stemming
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]  ## Removing Stopwords
    review = ' '.join(review)
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer


cv= CountVectorizer(max_features=2500)

x = cv.fit_transform(corpus).toarray()

y=df.iloc[:,4].values

print(x.shape,y.shape)
```

```python
# creating bag of words
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(max_features = 2500)

x = tf.fit_transform(corpus).toarray()
y = df.iloc[:, 4].values

print(x.shape)
print(y.shape)
```

```python
# creating bag of words
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(max_features = 2500,ngram_range=(2, 2))

x = tf.fit_transform(corpus).toarray()
y = df.iloc[:, 4].values

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

Xtrain,Xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
from sklearn.svm import SVC
svm=SVC(kernel="linear")

svm.fit(Xtrain,ytrain)
```

```python
ypredict=svm.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(ytest,ypredict)
```
